```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv('Movie.csv')
movies_df.head()
```

```python
movies_df.shape
#from the above dataframe there is not 8992 user ids, because one userid can watch other movie also and he can rate also
```

```python
len(movies_df.userId.unique())
```

```python
print(movies_df.movie.unique())
```

```python
#vector dimension = userId
#items(columns) = movies
user_movies_df = movies_df.pivot(index='userId',
                                columns='movie',
                                values='rating').reset_index(drop=True)
```

```python
user_movies_df
```

```python
user_movies_df.index = movies_df.userId.unique()
user_movies_df
```

```python
#Impute those NaNs with 0 values
user_movies_df.fillna(0, inplace=True)
user_movies_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_similarity = 1 - pairwise_distances(user_movies_df.values, metric='cosine')
```

```python
print(user_similarity)
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_similarity,)
user_sim_df
```

```python
#Set the index and colum names to user ids
user_sim_df.index = movies_df.userId.unique()
user_sim_df.columns = movies_df.userId.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
#here 3 and 11 are similar so we can recommand mutually
```

```python
np.fill_diagonal(user_similarity, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most similar Users
user_sim_df.idxmax(axis=1)[0:10]
#these two columns are almost similar so they can mutually recommend
```
